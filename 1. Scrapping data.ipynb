## 1. Scrapping GDP Data

## 2. Scrapping Unemployment Data

```python
import requests
import pandas as pd

api_key = "2884bccbd1cfaa9f5646c127814a97c0"
series_id = "LRUN64TTJPM156S"   # Japan Unemployment Rate (Monthly)

url = f"https://api.stlouisfed.org/fred/series/observations?series_id={series_id}&api_key={api_key}&file_type=json"

response = requests.get(url).json()


if "observations" not in response:
    raise ValueError("No data found. Check API key or series ID.")

df = pd.DataFrame(response['observations'])
df['value'] = pd.to_numeric(df['value'], errors='coerce')

print(df.head())

df.to_csv("japan_unemployment.csv", index=False)

print("Saved: japan_fred.csv")
```

## 3. Scrapping CPI Data

```python
import requests
import pandas as pd

api_key = "2884bccbd1cfaa9f5646c127814a97c0"
series_id = "JPNCPIALLMINMEI"   # Japan CPI All Items (Monthly)

url = (
    "https://api.stlouisfed.org/fred/series/observations"
    f"?series_id={series_id}&api_key={api_key}&file_type=json"
)

response = requests.get(url).json()


if "observations" not in response:
    raise ValueError("No CPI data returned. Check API key or series ID.")

df = pd.DataFrame(response["observations"])
df["value"] = pd.to_numeric(df["value"], errors="coerce")

print(df.head())
df.to_csv("japan_cpi_monthly.csv", index=False)

print("Saved: japan_cpi_headline_monthly.csv")
```

## 4. Scrapping Consumption Data

```python
import requests
import pandas as pd

api_key = "2884bccbd1cfaa9f5646c127814a97c0"
series_id = "NAEXKP01JPQ661S"   # Japan: Real Private Final Consumption

url = f"https://api.stlouisfed.org/fred/series/observations?series_id={series_id}&api_key={api_key}&file_type=json"

response = requests.get(url).json()

df = pd.DataFrame(response["observations"])  # This WILL exist
df["date"] = pd.to_datetime(df["date"])
df["value"] = pd.to_numeric(df["value"], errors="coerce")

df = df[["date", "value"]]
df.columns = ["Quarter", "RealConsumption"]

print(df.head())

```

```python
df.to_csv("japan_consumption.csv", index=False)

print("Saved: japan.csv")
```
